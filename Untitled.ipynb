```r
random_vector <- sample(2,50, replace=T, prob=c(.7,.3))
```

```r
# A separação é feita, a partir do vetor-aleatório
                                        # técnica usadas: predicados/condicionais e slicing/indexação
treino <- mymatrix[which(random_vector==1),] #posições de linha, tal que os valores são 1 no vertor amostra 1; para todas as colunas.
treino
```

```r
dados_controle <- mymatrix[which(random_vector==2),]
dados_controle
```

```r
## library(neuralnet)
n <- names(treino)
f <- as.formula(paste("Armas ~", paste(n[!n %in% "Armas"], collapse = " + ")))
nn <- neuralnet(f,data=treino,hidden=c(5,3),linear.output=T)
```
